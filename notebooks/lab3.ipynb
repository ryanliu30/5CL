```python
import sys
sys.path.append("..")
from utils import fit_and_plot, montecarlo_estimator
import numpy as np
import pandas as pd
```

```python
class ExpModel:
    def forward(self, m, D):
        return 2*m*650*1.22*6.35e-4/D
    def backward(self, m, D):
        return  2*m*650*1.22*6.35e-4/D
expmodel = ExpModel()
```

```python
m = np.arange(5) + 1
m_sig = m*np.sqrt(2)/6500
y = np.array([2.0, 3.6, 5.1, 6.6, 8.3])
y_sig = np.array([0.4]*5)
```

```python
fit_and_plot(
    expmodel,
    m, 
    y,
    m_sig,
    y_sig,
    1,
    None,
    "Order",
    "Diameter (mm)",
    "Circular Diffraction",
    r"$D = 2mL\frac{1.22\lambda}{D_{aperture}}$",
    True,
    );
```
