```python
class ExpModel:
    def forward(self, x, R):
        return R/4 - R*x
    def backward(self, x, R):
        return  -R
expmodel = ExpModel()
l = np.array([672, 490, 440])
l_sig = np.array([9, 10, 10])
lr = 1/l
lr_sig = l_sig/l**2
x = np.array([3, 4, 5])
x = 1/x**2
fit_and_plot(
    expmodel,
    x, 
    lr,
    np.zeros_like(x),
    lr_sig,
    1,
    None,
    r"$1/n_i^2$",
    r"1/\lambda (nm^{-1})",
    "Rydberg's Constant",
    r"$1/\lambda = R(\frac{1}{2^2} - \frac{1}{n_i^2})$",
    True,
    );
```
